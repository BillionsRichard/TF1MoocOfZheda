### 读取数据

```python
from sklearn.datasets import load_boston
```

```python
house_data = load_boston()
type(house_data)
```

```python
house_np = house_data.data
target_np = house_data.target
```

```python
house_np.shape, target_np.shape
```

```python
house_df = pd.DataFrame(house_np)
house_df.head()
```

```python
house_data.feature_names
```

```python
x = house_np.astype('float32')
y = target_np
```

```python
x.shape, y.shape, x.dtype
```

```python
import tensorflow as tf
```

```python
X = tf.placeholder(tf.float32, [None, 13], name='X')
Y = tf.placeholder(tf.float32, [None, 1], name='Y')
```

## 定义命名空间

```python
with tf.name_scope('Model'):
    w = tf.Variable(tf.random_normal([13, 1], stddev=0.01), name='W', dtype='float32')
    b = tf.Variable(1.0, name='b', dtype='float32')
    
    def model(x, w, b):
        return tf.matmul(x, w) + b
    
    pred = model(X, w, b)
    
```

```python
train_epochs = 50
learning_rate = 0.01
```

```python
with tf.name_scope('LossFunction'):
    loss = tf.reduce_mean(tf.pow(Y - pred, 2))
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)
```

```python
from random import shuffle
shuffle([1,2,3])
```

```python
for epoch in range(train_epochs):
    loss_sum = 0.0
    for sx, sy in zip(x, y):
        sx = sx.reshape(1, 13)
        sy = sy.reshape(1, 1)
        _,loss_ = sess.run([optimizer, loss], feed_dict={X:sx, Y:sy})
        loss_sum += loss_
    
    #xvalues, yvalues = shuffle(x, y)
    b0tmp = b.eval(session=sess)
    w0temp = w.eval(session=sess)
    loss_avg = loss_sum/len(y)
    print(f"Epoch:{epoch+1}, loss={loss_avg}, b={b0tmp}, w={w0temp}")
```
